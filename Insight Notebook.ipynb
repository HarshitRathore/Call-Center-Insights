```python
import pandas as pd
```

```python
df = pd.read_csv("january.txt", delimiter='\t')
```

```python
i = 0
for d in df['date']:
        x = pd.to_datetime(str(d),format='%y%m%d').strftime('%d-%m-%Y')
        print(df['date'][i],str(x))
        df['date'][i] = str(x)
        i = i + 1
```
